```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
os.chdir('/storage/emulated/0/Sales_Data')
```

```python
files=[file for file in os.listdir('/storage/emulated/0/Sales_Data')]
```

```python
all=pd.DataFrame()
for item in files:
      df=pd.read_csv(item)
      all=pd.concat([all,df])
    
              
 
```

```python
all['date']=all['Order Date'].str.split(' ').str[0]
```

```python
all['month']=all['date'].str.split('/').str[0]
```

```python
all
```

```python
(all.isna().mean())*100
```

```python
all.dropna(inplace=True)
```

```python
all=all.loc[all['month']!='Order']
```

```python
all['month']=all['month'].astype(int)
```

```python
all['Price Each']=pd.to_numeric(all['Price Each'])
```

```python
all['Quantity Ordered']=pd.to_numeric(all['Quantity Ordered'])
```

```python
all['Sales']=all['Price Each']*all['Quantity Ordered']
```

```python
all
```

### Question 1: What was the best month for sales? How much was earned that month?

```python
top_month=all.groupby('month')['Sales'].sum().sort_values(ascending=False).reset_index()
```

```python
# Define the custom palette
custom_palette = ["#008170", "#005B41"]
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.rcParams['axes.facecolor'] = 'lightgrey'
# Create the bar plot for month
sns.barplot(x='month', y='Sales', data=top_month, palette=custom_palette)
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.title('SALES DISTRIBUTION IN EACH MONTH OF YEAR',color='blue')
plt.tight_layout()
plt.show()
```

### Based on the analysis, the month with the highest sales is December. The bar plot showing the "SALES DISTRIBUTION IN EACH MONTH OF YEAR" clearly indicates that the sales peak in the 12th month, which is December.

### Question 2: What city sold the most product?

```python
all['city']=all['Purchase Address'].str.split(',').str[1]+' '+(all['Purchase Address'].str.split(',').str[2]).str.split(' ').str[1]
```

```python
top_city=all.groupby('city')['Sales'].sum().sort_values(ascending=False).reset_index()
top_city
```

```python
custom_palette = ["#008170", "#005B41"]  # Example colors, you can adjust as needed

# Set the style and background color
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.rcParams['axes.facecolor']='lightgrey'
sns.barplot(x='Sales', y='city', data=top_city, palette=custom_palette)
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_width():.0f}', (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha='left', va='center', fontsize=10, color='black', xytext=(5, 0),
                textcoords='offset points')
plt.title('SALES DISTRIBUTION IN EACH CITY OF USA',color='blue')
plt.tight_layout()
plt.show()
```

### According to the analysis, the city that sold the most products is San Francisco, CA. The bar plot showing the "SALES DISTRIBUTION IN EACH CITY OF USA" ranks San Francisco as the highest, with sales of $8,262,203.91.

### Question 3: What time should we display advertisements to maximize likelihood of customer’s buying product.

```python
all['Order Date']=pd.to_datetime(all['Order Date'])
```

```python
all['Hour']=all['Order Date'].dt.hour
```

```python
top_hour=all.groupby('Hour')['Sales'].sum().reset_index()
top_hour
```

```python
# Define the custom palette
custom_palette = ["#008170"]  # Example colors, you can adjust as needed

# Set the style and background color
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.rcParams['axes.facecolor'] = 'lightgrey'  # Change background color

# Assuming 'top_hour' is your DataFrame containing the data
# Make sure 'top_hour' contains 'Hour' column and 'Sales' column

# Create the line plot
sns.lineplot(x='Hour', y='Sales', data=top_hour, color ='darkgreen')

# Add title and adjust ticks
plt.title('Sales Distribution by Hour', fontsize=16, color='blue')
plt.xticks(top_hour['Hour'])  # Adjust x ticks
plt.show()
# Show the plot
plt.show()
```

### The line plot showing the "Sales Distribution Through Out The Day" suggests that the best time to display advertisements is around 11 AM to 2 PM. During this time window, the sales reach their peak, indicating a higher likelihood of customers making purchases.
